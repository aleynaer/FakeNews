```python
# gerekli kütüphaneler import edilir

import pandas as pd
import numpy as np
import nltk 

#nltk.download("stopwords")
from nltk.corpus import stopwords

#nltk.download("wordnet")
from nltk.stem import WordNetLemmatizer

from nltk.stem import PorterStemmer
```

```python
addr_f = r"C:\Users\Asus\Desktop\ml_project1\data\clean_fake.txt"

with open(addr_f) as f: 
    fake_headers = f.read().splitlines()
    f.close()

addr_r = r"C:\Users\Asus\Desktop\ml_project1\data\clean_real.txt"

with open(addr_r) as f: 
    real_headers = f.read().splitlines()
    f.close()

print(fake_headers)
```

```python
# okunan metinlerde nlp işlemleri uygula: kelimeleri tokenla,noktalama işaretlerini kaldıe ve  köklerine ayır

fake_corpus = []

for i in range(len(fake_headers)):
    temp_text = fake_headers[i]
    #print("\n ************** \n")
    #print(temp_text)
    
    tokenList = nltk.word_tokenize(temp_text)
    tokenList = list(filter(lambda token: nltk.tokenize.punkt.PunktToken(token).is_non_punct, tokenList))
    tokenList = list(filter(lambda token: token not in stopwords.words("english"), tokenList))
    
    lemmatizer = WordNetLemmatizer()
    tokenList = [lemmatizer.lemmatize(word) for word in tokenList]
    
    stemmer = PorterStemmer()
    tokenList = [stemmer.stem(word) for word in tokenList]

    tokenList = [word.lower() for word in tokenList]
    
    fake_corpus.append(tokenList)
    
fake_corpus
```

```python
# okunan metinlerde nlp işlemleri uygula: kelimeleri tokenla,noktalama işaretlerini kaldıe ve  köklerine ayır

real_corpus = []

for i in range(len(real_headers)):
    temp_text = real_headers[i]
    #print("\n ************** \n")
    #print(temp_text)
    
    tokenList = nltk.word_tokenize(temp_text)
    tokenList = list(filter(lambda token: nltk.tokenize.punkt.PunktToken(token).is_non_punct, tokenList))
    tokenList = list(filter(lambda token: token not in stopwords.words("english"), tokenList))
    
    lemmatizer = WordNetLemmatizer()
    tokenList = [lemmatizer.lemmatize(word) for word in tokenList]
    
    stemmer = PorterStemmer()
    tokenList = [stemmer.stem(word) for word in tokenList]

    tokenList = [word.lower() for word in tokenList]
    
    real_corpus.append(tokenList)
    
real_corpus
```

```python
fake_tokens = []

for i in range(len(fake_corpus)):
    sentence = fake_corpus[i]
    #print(sentence)
    for j in range(len(sentence)):
        fake_tokens.append(sentence[j])
        #words.values[j] = pd.DataFrame(tokens, columns = ["Words"])
        #print(token)
```

```python
real_tokens = []

for i in range(len(real_corpus)):
    sentence = real_corpus[i]
    #print(sentence)
    for j in range(len(sentence)):
        real_tokens.append(sentence[j])
        #words.values[j] = pd.DataFrame(tokens, columns = ["Words"])
        #print(token)
```

```python
fake_words = pd.DataFrame(fake_tokens, columns = ["fake_words"])
fake_words.sort_values("fake_words", ascending = True, inplace = True, ignore_index=True)
fake_words
```

```python
fake_words.drop_duplicates(inplace=True)
fake_words.set_index(np.arange(len(fake_words)), inplace = True)
fake_words
```

```python
real_words = pd.DataFrame(real_tokens, columns = ["real_words"])
real_words.sort_values("real_words", ascending = True, inplace = True, ignore_index=True)
real_words
```

```python
real_words.drop_duplicates(inplace=True)
real_words.set_index(np.arange(len(real_words)), inplace = True)
real_words
```

```python
corpus = fake_corpus + real_corpus
corpus
```

```python
news = fake_headers + real_headers
news
```

```python
df = pd.DataFrame(news, columns = ["headers"])
df
```

```python
# fakeFlag = 0 ise haber gerçek
# fakeFlag = 1 ise haber sahte

fakeFlag = np.zeros(len(news))
df["Fake_Flag"] = fakeFlag

for x in range(len(df)):  
    header = df["headers"].values[x] # sırayla header seçiyoruz
    flag = 0 # başlagıçta haber sahte değil
    
    for y in range(len(fake_headers)): # fake headerlerin içinde mi diye bakıyoruz
        isfake = fake_headers[y]
        if(header == isfake):
            flag = 1
            
    df["Fake_Flag"].values[x] = flag
    
df
```

```python
isFake = np.zeros(len(news))
df["isFake"] = isFake

isReal = np.zeros(len(news))
df["isReal"] = isReal

df
```

```python
df1 = df.iloc[:,2:4].copy()
df1
```

```python
# mevcut başlık içinde kaç tane fake haber corpusundan kelime var hesaplanır
# normalize etmek için bulunan sonuç cümle uzunluğuna bölünür (temizlenmiş cümle)

counter = 0

for x in range(len(corpus)): # satır satır (header) gez df'te
    sentence = corpus[x] # cümle seçtik
    for i in range(len(sentence)): # cümle içindeki kelimeleri gezer
        curr_word = sentence[i]
        if(curr_word in fake_words["fake_words"].values):
            #print("yes", curr_word)
            #print("***")
            counter += 1
    df["isFake"].values[x] = counter/len(sentence)
    counter = 0
```

```python
# mevcut başlık içinde kaç tane real haber corpusundan kelime var hesaplanır
# normalize etmek için bulunan sonuç cümle uzunluğuna bölünür (temizlenmiş cümle)

counter = 0

for x in range(len(corpus)): # satır satır (header) gez df'te
    sentence = corpus[x] # cümle seçtik
    for i in range(len(sentence)): # cümle içindeki kelimeleri gezer
        curr_word = sentence[i]
        if(curr_word in real_words["real_words"].values):
            #print("yes", curr_word)
            #print("***")
            counter += 1
    df["isReal"].values[x] = counter/len(sentence)
    counter = 0
```

```python
# train - test spliti

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.15, stratify=y,random_state=5)

# train - validation spliti

X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size = 0.174, stratify=y_train ,random_state=5)
```

```python
from sklearn.naive_bayes import BernoulliNB
from sklearn.naive_bayes import GaussianNB
from sklearn.naive_bayes import MultinomialNB
```

```python
# Multinomial NB ile model kurma

MultiNB_clf = MultinomialNB()

MultiNB_clf.fit(X_train, y_train)

preds = MultiNB_clf.predict(X_test)
score_MultiNB = MultiNB_clf.score(X_test, y_test)
print(score_MultiNB*100)
print()
print(classification_report(y_test, preds))

cm = confusion_matrix(y_test, preds)
sns
sns.heatmap(cm, annot=True,fmt='g')
```
